```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
amazon_reviews = pd.DataFrame({'Reviews':reviews,'Labels':labels})
```

```python
amazon_reviews.head()
```

```python
amazon_reviews['Labels']=amazon_reviews['Labels'].astype('int')
```

NLTK: Bag of Words Pipeline


*   Tokenization
*   Stopword Removal
*   Stemming
*   Vectorization





```python
amazon_rev=amazon_reviews['Reviews'].values.tolist()[:10000]
amazon_label=amazon_reviews['Labels'].values.tolist()[:10000]
```

```python
amazon_rev
```

```python
import nltk
nltk.download('punkt')
```

```python
#Tokenization
from nltk.tokenize import word_tokenize
tokenize_reviews = [word_tokenize(rev) for rev in amazon_rev]
```

```python
tokenize_reviews
```

```python
#2.Stopwords Removal
nltk.download('stopwords')
from nltk.corpus import stopwords
sw=stopwords.words('english')

cleaned_reviews=[]
for rev in tokenize_reviews:
  cleaned_rev=[word for word in rev if word not in sw]
  cleaned_reviews.append(cleaned_rev)
```

```python
cleaned_reviews[0]
```

```python
tokenize_reviews[0]
```

```python
#3. Lematizing
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer #---> PorterStemmer, SnowballLematizer, etc.
wnl= WordNetLemmatizer()
stemmed_reviews=[]
for rev in cleaned_reviews:
  stemmed_rev=[wnl.lemmatize(word) for word in rev]
  stemmed_reviews.append(stemmed_rev)
  
```

```python
stemmed_reviews[0]
```

```python
updated_reviews=[]
for rev in stemmed_reviews:
  updated_reviews.append(" ".join(rev))
```

```python
updated_reviews[0]
```

```python
#We are joining the words here to make the sentences so that we can give it to the count vectorizer or the inverse document frequency
```

#Inverse Document Frequency (IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect=TfidfVectorizer()
X= tfidf_vect.fit_transform(updated_reviews)
```

```python
X
```

```python
X_arr=X.toarray()
```

```python
X_arr.shape
```

```python
tfidf_vect.get_feature_names()
```

```python
tfidf_vect.vocabulary_
```

```python
np.unique(labels,return_counts=True)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,np.array(amazon_label),test_size=0.2,random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
 from sklearn.naive_bayes import MultinomialNB 
 mnb =MultinomialNB()
 #Fit the model
 mnb.fit(X_train,y_train)
 
```

```python
predictions=mnb.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix

#Accuracy
acc= accuracy_score(y_test,predictions)

#AUC Score
auc_score=roc_auc_score(y_test,predictions)

#Confusion Matrix

cnf_matrix=confusion_matrix(y_test,predictions)
```

```python
cnf_matrix
```

```python
import seaborn as sns
sns.heatmap(cnf_matrix,annot=True,cbar=False,xticklabels=['Negative Review','Positive Review'],yticklabels=['Negative Review','Positive Review'])
plt.yticks()
plt.title('Confusion Matrix')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.show()
```
