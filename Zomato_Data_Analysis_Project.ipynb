```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

zomato_data= pd.read_csv("/content/zomato.csv",encoding='latin-1')
```

```python
zomato_data.head()
```

```python
#Extract the Indian data
zomato_indian_data= zomato_data[zomato_data['Country Code']==1]
```

```python
zomato_indian_data
```

```python
zomato_indian_data['City'].value_counts()
```

```python
zomato_ncr= zomato_indian_data[(zomato_indian_data['City']=='New Delhi') | (zomato_indian_data['City']=='Gurgaon')| (zomato_indian_data['City']=='Noida')| (zomato_indian_data['City']=='Faridabad')]
```

```python
zomato_ncr
```

```python
zomato_ncr.shape
```

```python
zomato_ncr.drop(['Country Code','Currency'],axis=1,inplace=True)
```

```python
zomato_ncr.shape
```

```python
zomato_ncr.info() #to get the detailed information about each function
```

```python
zomato_ncr.isnull().sum() # to check the sum of count of null values in the data
```

```python
#Encoding the Values of Yes and NO to 1 and 0 respectively
colums =['Has Table booking','Has Online delivery','Is delivering now','Switch to order menu']
encoding={'Yes':1,'No':0}
for col in colums:
  zomato_ncr[col]=zomato_ncr[col].apply(lambda x : encoding[x])
```

```python
zomato_ncr
```

```python
pd.set_option('display.max_rows',None) # this is used to display all the rows in the expanded form
zomato_ncr['Restaurant Name'].value_counts()
```

```python
# To get the value count of the restaurants with more than 5 restaurants
rest_more_than_5={}
for restaurant, val in zip(zomato_ncr['Restaurant Name'].value_counts().index,zomato_ncr['Restaurant Name'].value_counts()):
  if val >=5:
    rest_more_than_5[restaurant]=val
print(rest_more_than_5)
print("The no of restuarants in delhi having more than 5 outlets are {}".format(len(rest_more_than_5)))
```

```python
#Ploting  the bar graph to show the no of restuarants in delhi with more than 5 restuarants
plt.figure(figsize=(22,22))
plt.bar(x=rest_more_than_5.keys(),height=rest_more_than_5.values())
plt.xlabel('Restuarants')
plt.ylabel('No of outlets')
plt.xticks(rotation=90)
plt.yticks(np.arange(0,100,2))
plt.grid()
plt.show()
```

```python
Above_5_restaurants= pd.DataFrame(None)
for column in list(rest_more_than_5.keys()):
  extract= zomato_ncr[zomato_ncr['Restaurant Name']==column]
  Above_5_restaurants= pd.concat((Above_5_restaurants,extract))
Above_5_restaurants.head()
type(Above_5_restaurants['City'].value_counts())
```

```python
import seaborn as sns

plt.figure(figsize=(12,9))
sns.countplot(Above_5_restaurants['City'],data =Above_5_restaurants,order=Above_5_restaurants['City'].value_counts().index)
plt.yticks(np.arange(0,950,50))
plt.show()
```

```python
 #Competitor data
 '''Competitor are those who are selling IceCream, Desert, North Indian'''
 competitor=set()
 for restaurant,cuisine in zip(delhidata['Restaurant Name'],delhidata['Cuisines']):
   if 'Ice Cream' in cuisine or 'Fast Food' in cuisine or 'Desert' in cuisine or 'North Indian' in cuisine:
     competitor.add(restaurant)
competitor
```

```python
# Preparing a dataframe out of this
competitor_data = pd.DataFrame(None)
for col in competitor:
  data =delhidata[delhidata['Restaurant Name']==col]
  competitor_data=pd.concat((competitor_data,data),axis=0)
competitor_data.head()
```

```python
mean_competitor_data= competitor_data.groupby('Restaurant Name').mean()[['Average Cost for two','Price range','Aggregate rating','Votes']]
mean_competitor_data.hist(figsize=(12,15))
plt.show()
```

```python
#Creating a Google Map Object
def create_googlemap(data, columns):
  gmap = gmplot.GoogleMapPlotter(28.7041, 77.1025, 20)
  for col in columns:
    latitude = data[data['Restaurant Name'] == col]['Latitude'].values
    longitude = data[data['Restaurant Name'] == col]['Longitude'].values
    gmap.scatter(latitude, longitude, color = 'red', size = 300, marker = False)
    gmap.draw("/content/map.html")
create_googlemap(competitor_data, competitor)
```
